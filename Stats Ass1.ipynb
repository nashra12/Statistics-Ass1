Q5. Explain the concept of levels of measurement and give an example of a variable for each level.
Ans-> Nominal: Categories without order

    Example: Types of animals (dog, cat, bird)

Ordinal: Categories with a meaningful order

    Example: Customer satisfaction rating (very dissatisfied to very satisfied)

Interval: Numeric scales with equal intervals, no true zero

    Example: Temperature in Celsius or Fahrenheit

Ratio: This is the highest level of measurement. It includes all the properties of the interval level, with the addition of a true zero point, which allows for the measurement of absolute magnitudes and ratios.


Q6. Why is it important to understand the level of measurement when analyzing data? Provide an
example to illustrate your answer.

Ans-> Understanding the level of measurement is crucial when analyzing data because it dictates which statistical techniques are appropriate and how data can be interpreted.

 If you have ordinal data such as customer satisfaction ratings (e.g., very dissatisfied, dissatisfied, neutral, satisfied, very satisfied), calculating the mean is not appropriate. Instead, you should use the median or mode to summarize central tendency and use non-parametric tests like the Mann-Whitney U test for analysis.
 When analyzing temperature data measured in Celsius (interval level), you can discuss the difference between temperatures (e.g., 30°C is 10°C warmer than 20°C)

Q8. Which type of plot can be used to display data in terms of range?
Ans->  A box plot (also known as a box-and-whisker plot) is an effective way to display data in terms of range. Box plots provide a visual summary of the distribution of a dataset, showing the central tendency, variability, and skewness of the data.


Q9. Describe the difference between descriptive and inferential statistics. Give an example of each
type of statistics and explain how they are used.
Ans-> Descriptive Statistics:

Descriptive statistics involve summarizing and organizing data to describe the characteristics of a dataset.
Example Scenario: A teacher calculates the average score of a class on a test.

    Data: Test scores of 30 students.
    Descriptive Statistics:
        Mean Score: Sum of all scores divided by the number of students.
        Median Score: The middle score when all scores are arranged in ascending order.
        Standard Deviation: A measure of how spread out the scores are around the mean
        
        Inferential Statistics:

Inferential statistics involve making predictions or inferences about a population based on a sample of data.
Example Scenario: A researcher wants to determine if a new teaching method is more effective than the traditional method.

    Data: Test scores of a sample of students taught using the new method and a sample taught using the traditional method.
